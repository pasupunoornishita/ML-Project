```python
#import libraries

import qrcode
import matplotlib.pyplot as plt
```

```python
#ADD data which you want to be QR Representation
data = "http://1.bp.blogspot.com/-Eg0woSZZRxw/VeribGKumKI/AAAAAAAABqo/XvGuN122hPM/s1600/happy_birthday_hd_images_12.jpg"
# Output file name

filename = "MY_QR_CODE.png"
#Generate qr code

image = qrcode.make(data)

#save imge in to a file

image.save(filename)
plt.imshow(image,cmap='gray')
```
